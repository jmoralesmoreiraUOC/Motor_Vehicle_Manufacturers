```python
# 1. Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#2. recurso de años de descaga
years = range(2000,2022)
id_table = [11, 10, 9, 8, 7, 3, 6, 13, 14, 15, 16, 25, 26, 27, 30, 33, 36, 40, 44, 46, 47, 48]
table_dict = {}

for i in range (0,len(years)):
    table_dict[years[i]] = id_table[i]
```

```python
# 3. implementation of web scraping
for k in table_dict:
    
    # Create an URL object
    url = 'https://www.oica.net/category/production-statistics/'+str(k)+'-statistics/'
    
    # Create object page
    page = requests.get(url)
    
    # parser-lxml = Change html to Python friendly format
    # Obtain page's information
    soup = BeautifulSoup(page.text, 'lxml')
    
    # Obtain information from tag <table>
    table_tag = soup.find('table', id='tablepress-'+str(table_dict[k]))
    
    # Create Dataframe only for the first year

    if (k == 2000):
        
        # Obtain every title of columns with tag <th>
        headers = []
        for i in table_tag.find_all('th'):
            if (len(headers) <=4 ):
                title = i.text
                headers.append(title)
        headers.append("Year")

        # Create a dataframe
        cars_DF = pd.DataFrame(columns = headers)
        
        # Create a for loop to fill mydata for the first year
        for j in table_tag.find_all('tr')[1:]:
            row_data = j.find_all('td')
            row = [i.text for i in row_data]
            length = len(cars_DF)
            row.append(k)
            if (len(row) > 1):
                cars_DF.loc[length] = row
        
        print("The year: " + str(k) + " ok")
    
    else:
        for j in table_tag.find_all('tr')[1:]:
            row_data = j.find_all('td')
            row = [i.text for i in row_data]
            length = len(cars_DF)
            row.append(k)
            if (len(row) > 1):
                cars_DF.loc[length] = row
            
        print("The year: " + str(k) + " ok")
        
cars_DF
```

```python
# 4. Exportar dataset al csv
from pathlib import Path
filepath = Path('C:/Cars_sales.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)
cars_DF.to_csv(filepath, index=False)

#cars_DF.to_csv(index=False)
```
